```python
# import modules
import pandas as pd
from bs4 import BeautifulSoup
from sqlalchemy import create_engine
```

```python
# locate the 1st Kaggle source
path = "static/Resources/The Guardian League Table Main Table 2013-2021.csv"
main_data = pd.read_csv(path)
main_data
```

```python
# locate the csv file exported from the web scraping code
path2 = "static/Resources/2022_data.csv"
scraped = pd.read_csv(path2)
scraped
```

```python
# check that the institution and year don't have nulls
main_data.isnull().any()
```

```python
# check that the institution and year don't have nulls
scraped.isnull().any()
```

```python
# Change a column name to match the actual column name from source
Fact = main_data.rename(columns = {
    "Career After 6 months" : "Career After 15 months"
})
Fact.head(15)
```

```python
# Create a dimension that would be used for filtering later on when visualizing.
Dim_uni = main_data [[
    "institution"
]]

Dim_uni_n = Dim_uni["institution"].unique()
Dim_uni_name = pd.DataFrame(Dim_uni_n, columns =['institution'])
Dim_uni_name
```

```python
# Append both files into one Fact table
Fact_data = scraped.append(Fact, ignore_index=True)
Fact_data
```

```python
Fact_data.to_excel("static/Resources/Combined.xlsx")
```

```python
# create engine that connect to the postgres sql db
protocol = 'postgresql'
username = 'postgres'
password = '258080'
host = 'localhost'
port = 5432
database_name = 'Uni_Ranking'
rds_connection_string = f'{protocol}://{username}:{password}@{host}:{port}/{database_name}'
engine = create_engine(rds_connection_string)
```

```python
#Check for the existing tables
engine.table_names()
```

```python
# pushing the Fact table to a created table in the db
Fact_data.to_sql(name='uni_ranking', con=engine, if_exists='append', index=False)
```

```python
# pushing the Dim table to a created table in the db
Dim_uni_name.to_sql(name='dim_uni_name', con=engine, if_exists='append', index=False)
```
